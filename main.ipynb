```python
from load_mnist import load_mnist
from model import train_model
from plot import training_curve_plot, plot_weights
```

### Load the dataset

```python
X_train, Y_train, X_test, Y_test = load_mnist()
```

### Evaluating a linear model (Neural Network with 0 hidden layers)

#### Using standard gradient descent

```python
K = 0 # Number of hidden layers
D = 0 # Number of units in each hidden layer
model_architecture = [K, D]
activation = 'ReLU'
optimizer = None
iterations = 600
learning_rate = 0.01
batch_size = 100

trained_weights, trained_biases, train_costs, test_costs, train_accuracies, test_accuracies = train_model(
    X_train, Y_train, model_architecture, activation, optimizer, iterations, learning_rate, batch_size,
    X_test, Y_test, print_every=50
)
```

```python
training_curve_plot('MNIST', train_costs, test_costs, train_accuracies, test_accuracies, batch_size, learning_rate)
```

```python
plot_weights(trained_weights[-1], "Output Layer Weights")
```
